# Thresholding

Thresholding is one of the most important and fundamental operations in computer vision.

Thresholding is often used to create binary images that allow you to perform follow-on processing tasks form many different types of applications.

---

+ In this notebook we are going to see **two types of thresholding operations**:
1. Global.
2. Adaptive.

---

This could be one of the first processing steps in an application that performs lane detection.

---

# 1. Global Thresholding

In OpenCV there is a threshold function and it has four required arguments
1. The image for which we want to perform thresholding on
2. Is a global threshold value that you specify.
3. The maximum value that's used for creating the binary map.
4. The type of thresholding that's going to be performed

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['image.cmap'] = 'gray'
```

```python
# Read the image.
img = cv2.imread('road_lanes.png', cv2.IMREAD_GRAYSCALE)

# Perform binary thresholding.
retval, img_thresh = cv2.threshold(img, 165, 255, cv2.THRESH_BINARY)

# Display the images.
plt.figure(figsize=(20, 8))
plt.subplot(121); plt.imshow(img); plt.title('Original')
plt.subplot(122); plt.imshow(img_thresh); plt.title('Thresholded');
```

# 2. Adaptive Thresholding

There is a function in opencv called **adaptiveThreshold** and it has six required arguments:
1. The grayscale image.
2. The max value for the binary map.
3. The method used for adaptive thresholding.
4. The threshold type.
5. (tuning parameter for the adaptive method).
6. (tuning parameter for the adaptive method).

Rather than specifying a single global threshold the adaptive method computes a threshold spatially  across the image.

```python
# Read the original image.
img = cv2.imread('Piano_Sheet_Music.png', cv2.IMREAD_GRAYSCALE)

# Perform global thresholding.
retval, img_thresh_gbl_1 = cv2.threshold(img, 50, 255, cv2.THRESH_BINARY)

# Perform global thresholding.
retval, img_thresh_glb_2 = cv2.threshold(img, 130, 255, cv2.THRESH_BINARY)

# Perform adaptive thresholding.
img_thresh_apd = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 13, 7)

# Display the images.
plt.figure(figsize=(18, 15))
plt.subplot(221); plt.imshow(img); plt.title('Original')
plt.subplot(222); plt.imshow(img_thresh_gbl_1); plt.title('Thresholded (global: 50)')
plt.subplot(223); plt.imshow(img_thresh_glb_2); plt.title('Thresholded (global: 130)')
plt.subplot(224); plt.imshow(img_thresh_apd); plt.title('Thresholded (adaptive)');
```
